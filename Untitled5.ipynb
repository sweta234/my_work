# Fill the missing values

```python
df1['max_salary']= df1['max_salary'].fillna(df1['max_salary'].mean())
df1['med_salary'] = df1['med_salary'].fillna(df1['med_salary'].mean())
df1['min_salary'] = df1['min_salary'].fillna(df1['min_salary'].mean())
df1['pay_period'] = df1['pay_period'].fillna('MONTHLY')
df1['applies'] = df1['applies'].fillna(df1['applies'].mean())
df1['remote_allowed'] = df1['remote_allowed'].fillna(1)
df1['views'] = df1['views'].fillna(df1['views'].mode())
df1['formatted_experience_level'] = df1['formatted_experience_level'].fillna('Entry level')
df1['currency'] = df1['currency'].fillna('USD')
df1['compensation_type'] = df1['compensation_type'].fillna('BASE_SALARY')
```

```python
df1['max_salary'] = df1['max_salary'].astype(int)
df1['med_salary'] = df1['med_salary'].astype(int)
df1['min_salary'] = df1['min_salary'].astype(int)
df1['applies'] = df1['applies'].astype(int)
```

```python
from sklearn.cluster import KMeans
```

```python
distortions = []
K = range(1, 11)
for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df3)
    distortions.append(kmeans.inertia_)

# Plot the elbow graph
plt.figure(figsize=(8, 5))
plt.plot(K, distortions, marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Distortion')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```
