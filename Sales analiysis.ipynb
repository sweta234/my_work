```python
data  = df[['MONDAY_ORDERS','TUESDAY_ORDERS','WEDNESDAY_ORDERS','THURSDAY_ORDERS','FRIDAY_ORDERS',
            'SATURDAY_ORDERS','SUNDAY_ORDERS','MONDAY_REVENUE','TUESDAY_REVENUE','WEDNESDAY_REVENUE','THURSDAY_REVENUE',
            'FRIDAY_REVENUE','SATURDAY_REVENUE','SUNDAY_REVENUE']]
```

```python
data.head(5)
```

```python
plt.subplots(figsize =(20,10))

sns.pointplot(x='MONDAY_ORDERS',y='MONDAY_REVENUE',data=data,color='black',alpha=0.8)
sns.pointplot(x='TUESDAY_ORDERS',y='TUESDAY_REVENUE',data=data,color='#CC0000',alpha=0.8)
sns.pointplot(x='WEDNESDAY_ORDERS',y='WEDNESDAY_REVENUE',data=data,color='#3399FF',alpha=0.8)
sns.pointplot(x='THURSDAY_ORDERS',y='THURSDAY_REVENUE',data=data,color='#3399FF',alpha=0.8)
sns.pointplot(x='FRIDAY_ORDERS',y='FRIDAY_REVENUE',data=data,color='#3399FF',alpha=0.8)
sns.pointplot(x='SATURDAY_ORDERS',y='SATURDAY_REVENUE',data=data,color='#3399FF',alpha=0.8)
sns.pointplot(x='SUNDAY_ORDERS',y='SUNDAY_REVENUE',data=data,color='#3399FF',alpha=0.8)

plt.xlabel('ORDER',fontsize = 15,color='blue')
plt.ylabel('REVENUE',fontsize = 15,color='blue')
plt.title('Solo vs Duo vs Squad Kills',fontsize = 20,color='blue')
plt.grid()
plt.show()
```
