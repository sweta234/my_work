```python
numeric_data = df.select_dtypes('int64')
```

```python
numeric_data
```

```python
categorical_data = df.select_dtypes('object')
```

```python
categorical_data
```

```python
num_data = df.select_dtypes('float64')
```

```python
num_data
```

```python
categorical_data.drop(['id','channel_sales'], axis =  1, inplace= True)
```

```python
categorical_data.drop(['origin_up'], axis =  1, inplace= True)
```

```python
categorical_data
```

```python
import datetime as dt

categorical_data['year_start_contract'] = pd.DatetimeIndex(categorical_data['date_activ']).year
categorical_data['year_end_contract'] = pd.DatetimeIndex(categorical_data['date_end']).year
```

```python
categorical_data
```

```python
categorical_data['contract_period_in year'] = categorical_data['year_end_contract'] - categorical_data['year_start_contract']
```

```python
categorical_data.drop(['date_activ','date_end','date_modif_prod','date_renewal'], axis =  1, inplace= True)
```

```python
categorical_data
```

```python
res = pd.concat([numeric_data , num_data,categorical_data], axis = 1)
```
